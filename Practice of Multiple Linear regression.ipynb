```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.graphics.correlation as sgc
import statsmodels.graphics.gofplots as qqplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.stats.api as sms
from sklearn import metrics
from statsmodels.stats.outliers_influence import OLSInfluence
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr
from sklearn.feature_selection import VarianceThreshold
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/Python/Crop_yield.csv")
df.head(5)
```

```python
X=df[['Temperature', 'Rainfall', 'Fertilizer_Usage', 'Pesticide_Usage']]
Y=df['Yield']
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2, random_state=42)
lm=LinearRegression()

modeled=lm.fit(X_train, Y_train)
prediction=modeled.predict(X_test)

mse=metrics.mean_squared_error(Y_test, prediction)
r2score=metrics.r2_score(Y_test, prediction)

model_save='crop_yield.pkl'
with open(model_save,'wb') as file:
    pickle.dump(modeled, file)

print(mse)
print(r2score)
print(model_save)
```

```python
loaded_model='crop_yield.pkl'
with open(loaded_model, 'rb') as file:
    loaded=pickle.load(file)

new_conditions = {
    'Temperature': [25],  # average temperature in °C
    'Rainfall': [150],  # total rainfall in mm
    'Fertilizer_Usage': [200],  # fertilizer used in kg per hectare
    'Pesticide_Usage': [30]  # pesticide used in litres per hectare
}
data=pd.DataFrame(new_conditions)

predicted_data=loaded.predict(data)
print(f"Predicted Yield for the new conditions: {predicted_data[0]} tonnes per hectare")
```
