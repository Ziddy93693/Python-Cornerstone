```python
class Tacco:
    def __init__(self, shell, protein, toppings=[]):
        self.shell=shell
        self.protein=protein
        self.toppings= toppings
        
    def __repr__(self):
        return (f'<Tacco toppings={self.toppings} protein={self.protein} shell={self.shell}'
                 + '@' + str(id(self))+ ">")
        
        
    def __str__(self):
        toppings= (f' with {",".join(self.toppings)}' if self.toppings else '')
        return (f'{self.shell}shell {self.protein} Tacco{toppings} ')

    def __format__(self,fmt_spec):
        if fmt_spec.endswith('d'):
            return str(sum(ord(char) for char in self.__str__()))
        if fmt_spec.endswith('Tacco'):
            return self.__repr__().upper() + '!!!!'
        
        return super().format__(fmt_spec)

    
        
t1= Tacco('hard','chicken')
t2= Tacco('soft', 'tofu', toppings=['cheese','lettuse','sour cream'])
```

```python
import sys
import time
class Buffer:
    def __init__(self, maxlength=50):
        self._data=b''
        self.maxlength=maxlength

    def write(self, msg):
        if len(self._data)< self.maxlength:
            self._data +=msg[:self.maxlength - len(self._data)]

    def flush(self):
        if self._data:
            sys.stdout.write(self._data.decode('utf-8'))
            self._data= b''

    def __del__(self):
        print("Deleting")
        self.flush()

b1=Buffer()
b1.write(b'Mike is hosting')
b2=Buffer()
b1.other =b2
del b1
while True:
    print("Stuff")
    time.sleep(1)
```

```python
class Taco:
    def __init__ (self, shell, toppings=[]):
        self.shell=shell
        self.toppings=toppings
        self.index=-1

    def __iter__(self):
        return (t for t in self.toppings)

    def __len__(self):
        return len(self.toppings)

    def __reversed__(self):
        for i in range(len(self)-1,-1,-1):
            yield self.toppings[i]
        

    
taco=Taco('Soft',['sour cream', 'tomatoes'])
```

```python
# Comparables
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def __repr__(self):
        return f'<point (x={self.x}, y={self.y})>'.format(self=self)

    def __lt__(self,other):
        self_dist=(self.x**2 + self.y**2)**0.5
        other_dist=(other.x**2 + other.y**2)**0.5
        return self_dist < other_dist 

    def __gt__(self,other):
        self_dist=(self.x**2 + self.y**2)**0.5
        other_dist=(other.x**2 + other.y**2)**0.5
        return self_dist > other_dist 
    
    def __le__(self,other):
        self_dist=(self.x**2 + self.y**2)**0.5
        other_dist=(other.x**2 + other.y**2)**0.5
        return self_dist <= other_dist 

    def __ge__(self,other):
        self_dist=(self.x**2 + self.y**2)**0.5
        other_dist=(other.x**2 + other.y**2)**0.5
        return self_dist >= other_dist 

    def __eq__(self, other):
        return self.x==other.x and self.y==other.y

    def __ne__ (self, other):
        return not self.__eq__(other)
       # return self.x!=other.x or self.y!=other.y

p1=point(0,0)
p2=point(0,0)
p3=point(1,-4)
p4=point(0,3)
        
```

```python
# Hashing and slots 
from random import randint
class car:
    def __init__(self, color, year):
        self.color=color
        self.year= year
        self.id=randint (0,10000)

    def __hash__(self):
        return hash((self.color, self.year, self.id))

    def __eq__(self,other):
        return (
            self.__class__==other.__class__ and 
            self.color == other.color and
            self.year == other.year and
            self.id == other.id
        )
        
first= car('red', 2000)
second= car('blue',2000)
third= car('red',2001)

reviews={}
reviews['first']='awesome'
reviews['second']='meh'
reviews['third']='it is good'
```

```python
# changing the color of car
def is_cars_equal(one, other):
    return (one.color == other.color) and (one.year == other.year)
```
