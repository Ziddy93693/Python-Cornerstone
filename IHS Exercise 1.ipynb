```python
# Bitwise
class binary:
    def __init__ (self, number):
        self.number=number
        self._binnumber=bin(number)
        
    def __repr__(self):
        return f'<Binary number={self.number} binnumber={self._binnumber}>'

    def __and__ (self, other):
        return binary(self.number & other.number)
        
    def __or__ (self, other):
        return binary(self.number | other.number)

    def __xor__ (self, other):
        return binary(self.number ^ other.number)

    def __rand__ (self,other):
        if isinstance (other, int):
            return binary(self.number & other)
        return binary(self.number & other.number)

    def __ror__ (self,other):
        if isinstance (other, int):
            return binary(self.number | other)
        return binary(self.number | other.number)

    def __rxor__ (self,other):
        if isinstance (other, int):
            return binary(self.number ^ other)
        return binary(self.number ^ other.number)

    def __iand__ (self, other):
        print('iand')
        temp=self.__and__(other)
        self.number=temp.number
        self._binnumber=temp.number
        return self

    def __ior__ (self, other):
        print('ior')
        temp=self.__or__(other)
        self.number=temp.number
        self._binnumber=temp.number
        return self

    def __ixor__ (self, other):
        print('ixor')
        temp=self.__xor__(other)
        self.number=temp.number
        self._binnumber=temp.number
        return self

b=binary(5)
b1= binary(7)
```

```python
from math import sin, cos, pi

class Vector3D:
    def __init__(self, x, y, z=0):
        self.x = x
        self.y = y
        self.z = z

    def __complex__(self):
        # Represent as complex using x and y
        return complex(self.x, self.y)


class Voltage:
    def __init__(self, base_voltage, frequency):
        self.base_voltage = base_voltage
        self.afrequency = 2 * pi * frequency

    def __complex__(self):
        # Default snapshot at t=0
        real = cos(self.afrequency * 0) * self.base_voltage
        imag = sin(self.afrequency * 0) * self.base_voltage
        return complex(real, imag)

    def at(self, t=0):
        real = cos(self.afrequency * t) * self.base_voltage
        imag = sin(self.afrequency * t) * self.base_voltage
        return complex(real, imag)
```

```python
# Clean up with context manager

class Car:
    def __init__(self):
        self.running =False
        self.odometer =0
        self.direction =90

    def move(self, amt):
        if self.running:
            self.odometer +=amt

    def _turn(self, amt):
        if self.running:
            self.direction +=amt

    def turn_right (self):
        self._turn(90)

    def turn_around(self):
        self._turn(180)

    def __enter__(self):
        self.running=True
        return self

    def __exit__(self, *args, **kwargs):
        self.running= False
        print(f'You have moved {self.odometer} miles')
        print(f'You are facing {self.direction} degrees')

with Car() as car:
    car.move(10)
    car.turn_right()
    car.move(10)
    car.turn_around()
```
