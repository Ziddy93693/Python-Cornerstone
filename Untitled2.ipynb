```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

df=pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/regression_sprint/regression_sprint_data_1.csv', index_col=0)
df.head(10)
```

```python
y=df['ZAR/USD']
X=df.drop('ZAR/USD',axis=1)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=50) 
```

```python
plt.scatter(X_train, y_train, color='green', label='Training')
plt.scatter(X_test, y_test, color='yellow', label='Testing')
plt.legend
plt.show()
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
predictions=lm.predict(X_train)
```

```python
a=lm.intercept_
b=float(lm.coef_[0])
print(a, b)
```

```python
plt.scatter(X_train,y_train, label='relationship')
plt.plot(X_train, predictions, color='red', label='line of best fit')
plt.title('The export vs USD')
plt.xlabel('The exports')
plt.legend
plt.ylabel('The USD')
plt.show()
```

```python
predictions1= lm.predict(X_test)
```

```python
plt.scatter(X_test,y_test, label='relationship')
plt.plot(X_test, predictions1, color='red', label='line of best fit')
plt.title('The export vs USD')
plt.xlabel('The exports')
plt.legend
plt.ylabel('The USD')
plt.show()
```

```python
from sklearn import metrics
```

```python
MSE_train= metrics.mean_squared_error(y_train, predictions)
MSE_test= metrics.mean_squared_error(y_test, predictions1)
RSS_train=(metrics.mean_squared_error(y_train, predictions))*len(X_train)
RSS_test=(metrics.mean_squared_error(y_test, predictions1))*len(X_test)
r2_Score_train=metrics.r2_score(y_train, predictions)
r2_Score_test=metrics.r2_score(y_test, predictions1)
```

```python
print(MSE_train
,MSE_test
,RSS_test
,RSS_train
,r2_Score_test
,r2_Score_train)
```
