```python
#Examine the following Python code intended to create a line plot to visualise the effect of temperature on maize yield. Identify any error/s in the code:

import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

data = pd.DataFrame({

    'Yield_of_maize_kgs': [250, 300, 350, 400, 450],

    'Average_temperature': [18, 30, 21, 25, 27]

})

sns.scatterplot(x='Yield_of_maize_kgs', y='Average_temperature', data=data)

plt.title('Effect of temperature on maize yield')

plt.show()
```
